```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
plt.plot(x,y)
```

```python
plt.plot(x,y,"-r")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```python
plt.subplot(1,2,2)
plt.plot(x,y,'r')
plt.subplot(2,2,1)
plt.plot(y,x,'b')
```

```python
# Object Oriented Programming with mtplotlib
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.set_title("Larger Figure")
axes2 = fig.add_axes([0.5,0.2,0.4,0.3])
axes2.set_title("Smaller figure")

axes1.plot(y,x,'r')
axes2.plot(x,y,'b')
```

```python
fig=plt.figure()
axe1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x,'r')
```
